```python
from tqdm import tqdm
import requests
from bs4 import BeautifulSoup
import re

import pandas as pd
import numpy as np
```

```python
# div = <div class="td-post-content tagdiv-type">
```

```python
df = pd.read_csv('/content/Input.xlsx - Sheet1.csv')
```

```python
all_urls =  []
for url in tqdm(df['URL']):

  try:
    # Send a GET request to the URL
    response = requests.get(url)

    soup = BeautifulSoup(response.content, 'html.parser')

    output = soup.find_all('div',class_  = "td-post-content tagdiv-type")

    result = output[0].text

    all_urls.append(result)

  except:
    all_urls.append(np.nan)
```

```python
df_part2 = pd.DataFrame(all_urls,columns=['extracted_text_from_urls'])
```

```python
final_data = pd.concat([df,df_part2],axis=1)
```
